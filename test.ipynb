# Imagenes

```python
import requests

url = "http://127.0.0.1:8000/upload-image/"
file_path = "input_files/dog.jpg"

with open(file_path, "rb") as file:
    response = requests.post(url, files={"file": file})

print(response.json())
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

url = "http://127.0.0.1:8000/get-image/dog.jpg"
response = requests.get(url)

if response.status_code == 200:
    image = Image.open(BytesIO(response.content))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
else:
    print("Failed to download image:", response.json())
```

```python
# URL of the FastAPI server endpoint
url = "http://127.0.0.1:8000/predict-image/"

# Path to the local image file you want to upload
file_path = "input_files/dog.jpg"

# Open the image file in binary mode
with open(file_path, "rb") as file:
    # Upload the file
    response = requests.post(url, files={"file": file})

# Print the server response
print(response.json())
```
